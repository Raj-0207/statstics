Random Sampling: A Simple Explanation
Random sampling is a method used to select individuals from a population in such a way that everyone has an equal chance of being chosen. This helps researchers make accurate conclusions about a larger group without studying every single person.

1. Why is Random Sampling Important?
âœ… Avoids Bias â€“ Ensures fairness by giving everyone an equal chance of selection.

Every nth person from a list is selected.
Example: Calling every 5th customer from a storeâ€™s contact list for feedback.

The population is divided into clusters (e.g., schools, cities), and entire clusters are chosen randomly.

#6. Explain the concept of skewness and its types. How does skewness affect the interpretation of data?

Skewness tells us how a dataset is distributedâ€”whether it's balanced or leaning more to one side. In simple terms, it shows if the data is spread out evenly or if thereâ€™s a long tail on one side.

Types of Skewness:
Positive Skew (Right-Skewed)

The tail on the right side is longer.
Most values are on the lower side, but a few high values pull the mean up.
Example: Income levelsâ€”most people earn a moderate amount, but a few earn very high salaries.
Negative Skew (Left-Skewed)

The tail on the left side is longer.
Most values are higher, but a few low ones pull the mean down.
Example: Exam scoresâ€”most students do well, but a few get very low marks.
Zero Skew (Symmetric Distribution)

Data is evenly spread on both sides of the mean.
The mean, median, and mode are nearly the same.
Example: Heights of adults in a population.
Why Does Skewness Matter?
Affects Averages: In skewed data, the mean gets pulled in the direction of the skew, so the median is often a better measure of the center.
Impacts Decision-Making: In finance or business, skewed data can highlight risks (like rare but extreme losses).
Statistical Assumptions: Many analysis methods assume normal (symmetric) data. If the data is skewed, you might need to transform it or use different statistical techniques.

#8. Discuss the conditions under which the binomial distribution is used

The binomial distribution is used when an experiment meets specific conditions and involves repeated independent trials with two possible outcomes.

Conditions for Using the Binomial Distribution:
Fixed Number of Trials (n):

The experiment is repeated a set number of times (e.g., flipping a coin 10 times).
Only Two Outcomes Per Trial:

Each trial has exactly two possible outcomes: success or failure (e.g., heads or tails, pass or fail).
Constant Probability of Success (p):

The probability of success remains the same for each trial (e.g., the chance of rolling a six on a fair die is always 1/6).
Independent Trials:

The outcome of one trial does not affect the outcome of another (e.g., drawing a card with replacement).
Examples of Binomial Distribution:
Tossing a coin multiple times and counting heads.
Checking if a manufactured product passes or fails a quality test.
Counting how many customers purchase a product in a sample of 50.

A random variable is a numerical value that represents the outcome of a random experiment. It assigns numbers to different possible results of a chance event.

Types of Random Variables:
Discrete Random Variable:

Takes only specific, countable values (like whole numbers).
